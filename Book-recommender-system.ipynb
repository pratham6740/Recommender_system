```python
ratings_with_name=ratings.merge(books,on='ISBN')
```

```python
ratings_with_name
```

```python
num_rating_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating': 'num_rating'},inplace=True)
num_rating_df
```

```python
pd.to_numeric(ratings_with_name['Book-Rating'], errors='coerce')
```

```python
ratings_with_name = ratings_with_name[(ratings_with_name['Book-Rating'] >= 0) & (ratings_with_name['Book-Rating'] <= 10)]
```

```python
ratings_with_name = ratings_with_name.dropna(subset=['Book-Rating'])
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean(numeric_only=True)['Book-Rating'].reset_index()
```

```python
avg_rating_df.rename(columns={'Book-Rating': 'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df= num_rating_df.merge(avg_rating_df, on='Book-Title')
popular_df
```

```python
popular_df[popular_df['num_rating']>=250]
```

```python
popular_df=popular_df[popular_df['num_rating']>=100].sort_values('avg_rating' , ascending=False )
```
